```python
# Create a copy
#INIT
df_cleaned = df.copy()
df_cleaned.shape
```

```python
#Removing Redundant variables
df_cleaned = df_cleaned.copy().drop(['Name','Ticket','Embarked', 'PassengerId'], axis=1)
df_cleaned.shape
```

```python
print(df_cleaned.Age.unique())
# print(df_cleaned.Embarked.unique())
```

```python
#Variable Selection Removing columns with Null Values
print(df_cleaned.shape)
NA_val = df_cleaned.isna().sum()
def na_filter(na, threshold = .4): #only select variables that passees the threshold
    col_pass = []
    for i in na.keys():
        if na[i]/df_cleaned.shape[0]<threshold:
            col_pass.append(i)
    return col_pass
print(df_cleaned.columns)
print(df_cleaned.shape)
df_cleaned = df_cleaned[na_filter(NA_val)] # Removing cabin
print(df_cleaned.columns)
df_cleaned.shape
```

```python
# Removing Outliers Not necessary in this case
```

```python
# Removing Rows with Null Values
df_cleaned = df_cleaned.dropna(axis=0)
df_cleaned.shape
```
